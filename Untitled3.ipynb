```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
fmri.describe()
```

```python
sns.lineplot(x="timepoint", y="signal", hue="event", palette=["teal", "indigo"], data=fmri)
plt.title("FMRI Signal Across Time")
plt.xlabel("Timepoint")
plt.ylabel("Signal Change")
```

```python
sns.relplot(x="timepoint", y="signal", col="region", hue="event", kind='line', data=fmri)
# kind can be left out and the default will be a scatter.
```

```python
plt.plot(fmri.timepoint, fmri.signal)
```

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.barplot(x="day", y="total_bill", data=tips)
# X and Y can be swapped over and sns automatically understands
# The lines that appear are confidence intervals. 
# If 'ci= None' is added then confidence intervals can be removed.
```

```python
plt.bar(tips.day, tips.total_bill)
```

```python
male = tips[tips['sex'] == 'Male']
female = tips[tips['sex'] == 'Female']

# Seperared into two different data frames

mday = male.groupby('day').mean()
fday = female.groupby('day').mean()


totals = mday.merge(fday, right_index =True, left_index=True)
totals['total'] = totals['tip_x'] + totals['tip_y']

totals['prop_m'] = np.true_divide(totals['tip_x'], totals['total']) * 100
totals['prop_f'] = np.true_divide(totals['tip_y'], totals['total']) * 100
totals
```

```python
plt.bar(totals.index, totals.prop_m, label='Male', color='gold', bottom=totals.prop_f)
# You can have a multiple stacked bar chart by inseting in the middle. 
# Last one does not have a bottom, bottom refers to the graph directly below.
plt.bar(totals.index, totals.prop_f, label='Female', color='green')
```
